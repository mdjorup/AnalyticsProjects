Malicious URL Features
- long length of url (int)
- random characters 
- number of dashes (int)
- look at www (binary)
- number of periods (int)
- number of percentage signs
- creation date (-1 more likely to be malicious)




```python

def get_url_lengths(url_lst):
  url_lengths = []
  for url in url_lst:
    url_lengths.append(len(url))
  return url_lengths

url_lengths = get_url_lengths(df['URL'])
print(url_lengths)

testdf['URL Lengths'] = get_url_lengths(testdf['URL'])




```

```python
df['URL Lengths'] = url_lengths
df
```

```python
def has_percent(url_list):
  has_percent = []
  for url in url_list:
    if "%" in url:
      has_percent.append(1)
    else:
      has_percent.append(0)
  return has_percent

percentage_sign = has_percent(df['URL'])
print(percentage_sign)

traindf['']
```

```python
def count_dashes(url_list):
  num_dashes = []
  for url in url_list:
    dash_count = url.count("-")
    num_dashes.append(dash_count)
  return num_dashes

num_dashes = count_dashes(df['URL'])
print(num_dashes)
testdf['Number of -'] = count_dashes(testdf['URL'])
```

```python
def count_percentages(url_list):
  num_percentages = []
  for url in url_list:
    percentage_count = url.count("%")
    num_percentages.append(percentage_count)
  return num_percentages

num_percentages = count_percentages(df['URL'])
print(num_percentages)
df['Number of %'] = num_percentages
testdf['Number of %'] = count_percentages(testdf['URL'])
```

```python
def count_periods(url_list):
  num_periods = []
  for url in url_list:
    periods_count = url.count(".")
    num_periods.append(periods_count)
  return num_periods

num_periods = count_periods(df['URL'])
print(num_periods)
df['Number of .'] = num_periods
testdf['Number of .'] = count_periods(testdf['URL'])
```

```python
def get_hostname(url):
  return url.split("://")[1].split("/")[0]


df["Hostname"] = df["URL"].apply(get_hostname)

def get_path(url):
  try:
    return url.split("://")[1].split("/")[1]
  except:
    return "..."

df["Path"] = df["URL"].apply(get_path)
testdf["Path"] = testdf["URL"].apply(get_path)

def count_slashes(path):
  return path.count("/");

df['Number of /'] = df["Path"].apply(count_slashes)

testdf['Number of /'] = testdf["Path"].apply(count_slashes)


```

```python
df
```

```python
def path_randomness(path):
  if(len(path) == 0):
    return 0
  def is_number(char):
      if ord(char) >= 48 and ord(char) <= 57:
          return (True)
      else:
          return (False)
  state = is_number(path[0])
  count = 0
  for char in path:
      if state is not is_number(char):
          count += 1
          state = not state
  return (float(count / len(path)))

  


df["Randomness"] = df["Path"].apply(path_randomness)
testdf["Randomness"] = testdf["Path"].apply(path_randomness)
```

```python
df
```

```python
def get_creation_indicator(url_list):
  url_indicators = []
  for val in url_list.itertuples():
    if -1 in val:
      url_indicators.append(1)
    else: url_indicators.append(0)
  return url_indicators

url_list = get_creation_indicator(df[['create_age(months)', 'expiry_age(months)', 'update_age(days)']])
url_list2 = get_creation_indicator(testdf[['create_age(months)', 'expiry_age(months)', 'update_age(days)']])

print(url_list)
df['Creation Indicator'] = url_list
testdf['Creation Indicator'] = url_list2
testdf
```
